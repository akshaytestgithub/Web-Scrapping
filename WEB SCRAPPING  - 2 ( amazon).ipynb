```python
from autoscraper import AutoScraper
```

```python
url = 'https://www.amazon.in/s?k=neckband+earpods'
wanted_list = ['₹1,299', 'Boult Audio ZCharge Bluetooth Wireless in Ear Earphones with Mic, 40H Playtime and Super Fast Charging, Environmental Noise Cancellation for Pro+ Calling and IPX5 Water Resistant (Black)']
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped = True)
```

```python
# taking titles out
```

```python
scraper.set_rule_aliases({'rule_buei':'Title', 'rule_8vsu':'Price'})
scraper.keep_rules(['rule_buei', 'rule_8vsu'])
scraper.save('amazon-earpod-search')
```

```python
results = scraper.get_result_similar('https://www.amazon.in/s?k=neckband+earpods', group_by_alias = True)
```
